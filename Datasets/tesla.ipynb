#### **Checking The Distribution Of the Data:**

```python
# Set style and color palette using Seaborn
sns.set_style("whitegrid")
palette = sns.color_palette("husl", 8)  # Using the 'husl' color palette for vibrant colors

# Plot using Seaborn with vibrant colors
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Open', palette=palette)
plt.title('Tesla Stock Open Prices Over Time', fontsize=16, fontweight='bold', color='blue')  # Adjust title properties
plt.xlabel('Date', fontsize=16, fontweight='bold', color='blue')  # Adjust x-axis label properties
plt.ylabel('Open Price', fontsize=16, fontweight='bold', color='blue')  # Adjust y-axis label properties
plt.xticks(rotation=45)
plt.grid(True)  # Add grid for better visualization
plt.show()
 
    # Plot the distribution of closing prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Open'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Tesla Stock openiing Prices', fontsize=16)
plt.xlabel('opening Price', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True)
plt.show()
```
